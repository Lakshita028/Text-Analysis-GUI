```python
# function to print sentiments
# of the sentence.
def detect_sentiment():
 
    # get a whole input content from text box
    sentence = load_f.get()
 
    # Create a SentimentIntensityAnalyzer object.
    sid_obj = SentimentIntensityAnalyzer()
 
    # polarity_scores method of SentimentIntensityAnalyzer
    # object gives a sentiment dictionary.
    # which contains pos, neg, neu, and compound scores.
    sentiment_dict = sid_obj.polarity_scores(sentence)
 
    string = str(sentiment_dict['neg']*100) + "% Negative"
    negativeField.insert(10, string)
     
 
    string = str(sentiment_dict['neu']*100) + "% Neutral"
    neutralField.insert(10, string)
 
    string = str(sentiment_dict['pos']*100) +"% Positive"
    positiveField.insert(10, string)
     
    # decide sentiment as positive, negative and neutral
    if sentiment_dict['compound'] >= 0.05 :
        string = "Positive"
 
    elif sentiment_dict['compound'] <= - 0.05 :
        string = "Negative"
      
 
    else :
        string = "Neutral"
 
    overallField.insert(10, string)
```

```python
# Building the GUI
root = tk.Tk()
root.title('ANALYZE YOUR TEXT!')
root.geometry('900x500+10+10')

# TAB Label
tab_control = ttk.Notebook(root)
home_tab = ttk.Frame(tab_control)
tab2 = ttk.Frame(tab_control)
tab1 = ttk.Frame(tab_control)

bgimg= tk.PhotoImage(file = "/Users/lakshitasethi/Downloads/img11.ppm")

# ADD TAB TO NOTEBOOK
tab_control.add(home_tab, text = "Home")
tab_control.add(tab2,text = "Sentiment Analysis")
tab_control.add(tab1,text = "Word Cloud Generator")

tab_control.pack(expand = 1, fill = "both")

# ADD LABEL TO HOME TAB
label = tk.Label(home_tab, text = """TEXT ANALYZER\n\n\nGetting trouble reading HUGE texts?\n Here's what you can try.""",font = ("Lunida 13",40),anchor = CENTER,image = bgimg, compound='center')
label.pack(padx = 0,pady = 0)

class CreateToolTip(object):
    """
    create a tooltip for a given widget
    """
    def __init__(self, widget, text='widget info'):
        self.waittime = 100     #miliseconds
        self.wraplength = 280   #pixels
        self.widget = widget
        self.text = text
        self.widget.bind("<Enter>", self.enter)
        self.widget.bind("<Leave>", self.leave)
        self.widget.bind("<ButtonPress>", self.leave)
        self.id = None
        self.tw = None

    def enter(self, event=None):
        self.schedule()

    def leave(self, event=None):
        self.unschedule()
        self.hidetip()

    def schedule(self):
        self.unschedule()
        self.id = self.widget.after(self.waittime, self.showtip)

    def unschedule(self):
        id = self.id
        self.id = None
        if id:
            self.widget.after_cancel(id)

    def showtip(self, event=None):
        x = y = 0
        x, y, cx, cy = self.widget.bbox("insert")
        x += self.widget.winfo_rootx() + 25
        y += self.widget.winfo_rooty() + 20
        # creates a toplevel window
        self.tw = tk.Toplevel(self.widget)
        # Leaves only the label and removes the app window
        self.tw.wm_overrideredirect(True)
        self.tw.wm_geometry("+%d+%d" % (x, y))
        label = tk.Label(self.tw, text=self.text, justify='left',
                       background="#ffffff", relief='solid', borderwidth=1,
                       wraplength = self.wraplength)
        label.pack(ipadx=1)

    def hidetip(self):
        tw = self.tw
        self.tw= None
        if tw:
            tw.destroy()

# LABELS FOR TAB1
label0=tk.Label(tab1, text="Hover over line for instructions")
label0.grid(row=0)
label1=tk.Label(tab1, text="Load File")
label1.grid(row=1)
label1_tooltip=CreateToolTip(label1,'Load a text file containing your words')
label2=tk.Label(tab1, text="Words")
label2.grid(row=2)
label2_tooltip=CreateToolTip(label2,'Type in the words you want in your wordcloud, the more words, the better the word cloud will look. \n Words are case-sensitive (i.e. The is different form the)\n Separate words with space')
label3=tk.Label(tab1, text="Frequency of word")
label3.grid(row=3)
label3=CreateToolTip(label3,'This signifies the size of the above words. The more frequent a word, the larger it will be.\n Type in a number that represents the size of the word \n'
'The larger the number, the larger the word. I.E. Family Love Health in order of size, would be 200 100 50')
label4=tk.Label(tab1, text="Background Color")
label4.grid(row=4)
label4=CreateToolTip(label4,'Background Color of the Word Cloud, Default is white')
label5=tk.Label(tab1, text="Wordcloud Width")
label5.grid(row=5)
label5=CreateToolTip(label5,'The higher this number, the better the resolution of the words within your word cloud. Default is 400')
label6=tk.Label(tab1, text="Wordcloud Height")
label6.grid(row=6)
label6=CreateToolTip(label6,'Same as the width. Default is 400')
label7=tk.Label(tab1, text="Window Width")
label7.grid(row=7)
label7=CreateToolTip(label7,'Size of the WordCloud picture. \nThe higher the number, the larger the figure. Default is 5')
label8=tk.Label(tab1, text="Window Height")
label8.grid(row=8)
label8=CreateToolTip(label8,'Same as width. Default is 5')
label9=tk.Label(tab1, text="Words not to use")
label9.grid(row=9)
label9=CreateToolTip(label9,'A list of words to ignore. \nSeparate multiple words with space. I.E. Hello Goodbye.\nWords are case-sensitive (i.e. The is different form the)')
tk.Label(tab1, text="Image Options").grid(row=10)
label10=tk.Label(tab1, text="Load Image")
label10.grid(row=11)
label10=CreateToolTip(label10,'Image to insert the wordcloud into. \n Use photos that have a white background, and black filling where you want the words to be in.')
label11=tk.Label(tab1, text="Image Outline Thickness")
label11.grid(row=12)
label11=CreateToolTip(label11,'This defines the outline of the image. Default is 3')
label12=tk.Label(tab1, text="Image Outline Color")
label12.grid(row=13)
label12=CreateToolTip(label12,'The color of the outline. Default is black')


# LABELS FOR TAB2
enterText = tk.Label(tab2, text="Enter the Text")
enterText.grid(row = 1, column = 1)

load_f= tk.Entry(tab2)
load_f.grid(row=1, column=2,ipadx = 100,sticky=W)

# create a Submit Button and place into the root window
# when user press the button, the command or 
# function affiliated to that button is executed 
check = Button(tab2, text = "Check Sentiment", fg = "Black", bg = "Red", command = detect_sentiment)
check.grid(row = 3, column = 2)

# Create a negative : label
negative = Label(tab2, text = "Sentence was rated as: ")
negative.grid(row = 4, column = 1)

# Create a neutral : label
neutral = Label(tab2, text = "Sentence was rated as: ")
neutral.grid(row = 6, column = 1)

# Create a positive : label
positive = Label(tab2, text = "Sentence was rated as: ")
positive.grid(row = 8, column = 1)

# Create a overall : label
overall = Label(tab2, text = "Sentence Overall Rated As: ")
overall.grid(row = 10, column = 1)

# create a text entry box 
negativeField = Entry(tab2)
negativeField.grid(row = 4, column = 2,ipadx = 100)

# create a text entry box 
neutralField = Entry(tab2)
neutralField.grid(row = 6, column = 2,ipadx = 100)

# create a text entry box 
positiveField = Entry(tab2)
positiveField.grid(row = 8, column = 2,ipadx = 100)

# create a text entry box 
overallField = Entry(tab2)
overallField.grid(row = 10, column = 2,ipadx = 100)

# ENTERING FILES/DATA IN TAB1
load_file=tk.Entry(tab1)
load_file.grid(row=1, column=1,ipadx = 100,sticky=W)
word_list=tk.Entry(tab1)
word_list.grid(row=2, column=1,ipadx=100,sticky=W+E)
multiplyer=tk.Entry(tab1)
multiplyer.grid(row=3, column=1,ipadx=100,sticky=W+E)
background_color_input=tk.Entry(tab1)
background_color_input.grid(row=4,column=1,ipadx = 100,sticky=W)
width_input=tk.Entry(tab1)
width_input.grid(row=5,column=1,ipadx=100,sticky=W)
height_input=tk.Entry(tab1)
height_input.grid(row=6,column=1,ipadx=100,sticky=W)
window_width_input=tk.Entry(tab1)
window_width_input.grid(row=7,column=1,ipadx=100,sticky=W)
window_height_input=tk.Entry(tab1)
window_height_input.grid(row=8,column=1,ipadx=100,sticky=W)
stopwords_input=tk.Entry(tab1)
stopwords_input.grid(row=9,column=1,ipadx=100,sticky=W)
image_input=tk.Entry(tab1)
image_input.grid(row=11,column=1,ipadx=100,sticky=W)
contour_width_input=tk.Entry(tab1)
contour_width_input.grid(row=12,column=1,ipadx=100,sticky=W)
contour_color_input=tk.Entry(tab1)
contour_color_input.grid(row=13,column=1,ipadx=100,sticky=W)

text=()
text_directory=()
image=()
image_directory=()

def input_file():
    fullpath = filedialog.askopenfilename(parent=tab1, title='Choose a file')
    global text_directory
    global text
    text_directory=os.path.dirname(fullpath)
    text= os.path.basename(fullpath)
    label2 = Label(tab1,text=fullpath).grid(row=1,column=1,sticky=W)
    
def input_file_2():
    fullpath = filedialog.askopenfilename(parent=tab2, title='Choose a file')
    global text_directory
    global text
    text_directory=os.path.dirname(fullpath)
    text= os.path.basename(fullpath)
    label2 = Label(tab2,text=fullpath).grid(row=1,column=1,sticky=W)

def image_file():
    fullpath = filedialog.askopenfilename(parent=tab1, title='Choose a file')
    global image
    global image_directory
    image_directory=os.path.dirname(fullpath)
    image= os.path.basename(fullpath)
    label2=Label(tab1,text=fullpath).grid(row=11,column=1,sticky=W)
    
def foo():      
    img1 = ImageTk.PhotoImage(img.open("/Users/lakshitasethi/Downloads/savefile.png").resize((400,300)))
    label = tk.Label(image = img1)
    label.image = img1
    label.place(x = 250, y = 520)
    
new_top_flag=False
new_top=None

def run():
    print('Program is Running')
    try:
        global new_top
        stopwords = set(STOPWORDS)
        width_value,height_value,background_colors,contour_color,contour_width,stopwords,window_width,window_height = defaults(stopwords)
        if new_top_flag != False:
            max_words,min_font_size,max_font_size,mode,relative_scaling=new_top.values()
        else:
            max_words,min_font_size,max_font_size,mode,relative_scaling=200,4,None,'RGB','auto'
        if word_list.get() != '':
            inputed_words=[]
            words_typed=word_list.get()
            multiplyer_value=multiplyer.get()
            for words,number in zip(words_typed.split(),multiplyer_value.split()):
                inputed_words+=((words+',')*int(number)).split(',')
            word_could_dict=Counter(inputed_words)
            if image == ():
                wc= WordCloud(width=width_value, height=height_value,background_color=background_colors,stopwords=stopwords,max_words=max_words,min_font_size=min_font_size,max_font_size=max_font_size,mode=mode,relative_scaling=relative_scaling).generate_from_frequencies(word_could_dict)
            else:
                os.chdir(image_directory)
                new_mask=np.array(img.open(image))
                wc= WordCloud(width=width_value, height=height_value,background_color=background_colors,stopwords=stopwords,mask=new_mask,contour_color=contour_color,contour_width=contour_width,max_words=max_words,min_font_size=min_font_size,max_font_size=max_font_size,mode=mode,relative_scaling=relative_scaling).generate_from_frequencies(word_could_dict)
            wc.to_file('savefile.png')
            
            plt.figure(figsize = (window_width,window_height))
            plt.imshow(wc, interpolation="bilinear")
            plt.tight_layout(pad=0)
            plt.axis("off")
            plt.show()
        else:
            os.chdir(text_directory)
            text_file = (open(text,encoding="utf8" ).read()).lower()
            if image ==():
                wc= WordCloud(width=width_value, height=height_value,background_color=background_colors,stopwords=stopwords,max_words=max_words,min_font_size=min_font_size,max_font_size=max_font_size,mode=mode,relative_scaling=relative_scaling).generate(text_file)
            else:
                os.chdir(image_directory)
                new_mask=np.array(img.open(image))
                wc= WordCloud(width=width_value, height=height_value,background_color=background_colors,stopwords=stopwords,mask=new_mask,contour_color=contour_color,contour_width=contour_width,max_words=max_words,min_font_size=min_font_size,max_font_size=max_font_size,mode=mode,relative_scaling=relative_scaling).generate(text_file)
            wc.to_file('savefile.png')
            plt.figure(figsize = (window_width,window_height))
            plt.imshow(wc, interpolation="bilinear")
            plt.tight_layout(pad=0)
            plt.axis("off")
            plt.show()
        new_top=None
    except:
        print('Error Running Program. Report below error report\n')
        traceback.print_exc()

# function to print sentiments
# of the sentence.
def detect_sentiment():
 
    # get a whole input content from text box
    sentence = load_f.get()
 
    # Create a SentimentIntensityAnalyzer object.
    sid_obj = SentimentIntensityAnalyzer()
 
    # polarity_scores method of SentimentIntensityAnalyzer
    # object gives a sentiment dictionary.
    # which contains pos, neg, neu, and compound scores.
    sentiment_dict = sid_obj.polarity_scores(sentence)
 
    string = str(sentiment_dict['neg']*100) + "% Negative"
    negativeField.insert(10, string)
     
 
    string = str(sentiment_dict['neu']*100) + "% Neutral"
    neutralField.insert(10, string)
 
    string = str(sentiment_dict['pos']*100) +"% Positive"
    positiveField.insert(10, string)
     
    # decide sentiment as positive, negative and neutral
    if sentiment_dict['compound'] >= 0.05 :
        string = "Positive"
 
    elif sentiment_dict['compound'] <= - 0.05 :
        string = "Negative"
      
 
    else :
        string = "Neutral"
 
    overallField.insert(10, string)
    
def defaults(stopwords):
    if width_input.get() == '':
        width_value=400
    else:
        width_value=int(width_input.get())
    if height_input.get() == '':
        height_value=400
    else:
        height_value=int(height_input.get())
    if stopwords_input.get() != '':
        for values in (stopwords_input.get()).split():
            stopwords.add(values)
    if background_color_input.get() == '':
        background_colors='white'
    else:
        background_colors=background_color_input.get()
    if contour_color_input.get() == '':
        contour_color='black'
    else:
        contour_color=contour_color_input.get()
    if contour_width_input.get() == '':
        contour_width=3
    else:
        contour_width=int(contour_width_input.get())
    if window_width_input.get() == '':
        window_width = 5
    else:
        window_width=int(window_width_input.get())
    if window_height_input.get() == '':
        window_height=5
    else:
        window_height=int(window_height_input.get())
    return width_value,height_value,background_colors,contour_color,contour_width,stopwords,window_width,window_height

def new_window():
    global new_top
    global new_top_flag
    from advanced import newTopLevel
    new_top = newTopLevel(root)
    new_top_flag=True
    newWindow = new_top.newWindow

# Function for clearing the  
# contents of all entry boxes
# And text area.
def clearAll() :
   
    # deleting the content from the entry box
    negativeField.delete(0, END)
    neutralField.delete(0, END)
    positiveField.delete(0, END)
    overallField.delete(0, END)
 
    # whole content of text area  is deleted
    load_f.delete(1.0, END)


# Buttons for Tab1
tk.Button(tab1,text='browse',command=input_file).grid(row=1,column=2)
tk.Button(tab1,text='browse',command=image_file).grid(row=11,column=2)
tk.Button(tab1,text='run',command=run).grid(row=14,column=1,sticky=W)
tk.Button(tab1,text = 'Show Word Cloud',command = foo).grid(row = 40,column = 1,sticky=W)

# Buttons for Tab2
# create a Clear Button and place into the root window
# when user press the button, the command or 
# function affiliated to that button is executed .
tk.Button(tab2, text = "Clear", fg = "Black", bg = "Red", command = clearAll).grid(row = 13, column = 1)

root.mainloop()
```
